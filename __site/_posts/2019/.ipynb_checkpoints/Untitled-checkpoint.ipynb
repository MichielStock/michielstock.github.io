```julia
using EcologicalNetworks
```

```julia
nsims = 10

αs = zeros(nsims)
βs = zeros(nsims)

ninteractions = zeros(Int, nsims)
resiliences = zeros(nsims)
heterogeneities = zeros(nsims)
symmetries = zeros(nsims)
I = zeros(nsims)
D = zeros(nsims)
V = zeros(nsims)

for i in 1:nsims
    α = rand()
    β = rand()
    
    P = [1 α; 0 β]
    #P = rand(2, 2)
    αs[i] = α
    βs[i] = β
    
    P10 = P ⊗ 10
    G = fastsample(P10)
    N = UnipartiteNetwork(G)
    ninteractions[i] = sum(N)
    resiliences[i] = resilience(N)
    heterogeneities[i] = heterogeneity(N)
    symmetries[i] = symmetry(N)
    itdec = information_decomposition(N)
    
    #nestedness[i] = η(BipartiteNetwork(G))
end
```

```julia
using DataFrames

DataFrame(Any[αs, βs, ninteractions, resiliences, heterogeneities, symmetries, nestedness],
            [:α, :β, :interactions, :resilience, :heterogeneity, :symmetry, :nestedness])
```
